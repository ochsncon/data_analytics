# Importing and preparing supermarkets data

## Libraries and settings

```python
# Libraries
import os
import fnmatch
import pandas as pd

# Ignore warnings
import warnings
warnings.filterwarnings("ignore")

# Get current working directory
print('Current working directory:', os.getcwd())

# Show .json files in the current working directory
flist = fnmatch.filter(os.listdir('.'), '*.json')
for i in flist:
    print(i)
```

## Importing data

```python
# Read the data to a pandas data frame
df1 = pd.read_json('supermarkets.json', encoding='utf-8')
df1.head(5)
```

## Count number of rows and columns in the data frame

```python
# Dimension (rows, columns)
print('Dimension:', df1.shape)

# Number of rows
print('Number of rows:', df1.shape[0])

# Number of columns
print('Number of columns:', df1.shape[1])
```

## Column 'tags' is a pandas Series with dictionaries -> change to data frame

```python
# Type of the first item of column tags
print(type(df1.tags))
print(type(df1.tags[0]))

# Content of the first item of column tags
print(df1.tags[0].keys())

# Change to data frame
df2 = pd.DataFrame.from_records(df1.tags)
df2 = df2[['brand', 'shop', 'addr:city', 'addr:street', 'addr:housenumber', 'addr:postcode']]

# Rename selected columns
df2 = df2.rename(columns={'addr:city': 'city',
                          'addr:street':'street',
                          'addr:housenumber': 'housenumber',
                          'addr:postcode': 'postcode'})

# Show first records of data frame
df2.head()
```

## Merge df1 and df2

```python
# Merge df and df2
df = pd.merge(df1[['type', 'id', 'lat', 'lon']], 
              df2[['brand', 'shop', 'city', 'street', 'housenumber', 'postcode']],
              left_index=True, 
              right_index=True)
df.head(5)
```

## Count and identify the number of missing values (if any)

## Count and identify duplicated values (if any)

```python
# Count duplicated values
print(df.duplicated().sum())

# Identify rows with duplicated values, e.g.:
df[df[['id']].duplicated()]
```

## Additional filters for supermarkets

```python
# e
import unicodedata
df_filtered = df.loc[(df['brand'] == 'Coop') & (df['city'] == 'Zürich')]
```

```python
# f
def norm_city(s):
    if pd.isna(s):
        return ''
    s = str(s).strip()
    # Diakritika entfernen (Zürich -> Zurich), dann kleinschreiben
    s = ''.join(c for c in unicodedata.normalize(
        'NFKD', s) if not unicodedata.combining(c))
    return s.lower()


city_norm = df['city'].map(norm_city)
is_zrh = city_norm.eq('zurich')
is_mig = df['brand'].fillna('').str.contains(
    r'\bmigros\b', case=False, na=False)

df_migros_zuerich = df.loc[is_zrh & is_mig].copy()
df_migros_zuerich.head()
```

```python
# g
city_norm = df['city'].map(norm_city)
target_cities = {'zurich': 'Zürich', 'basel': 'Basel', 'bern': 'Bern'}

is_target_city = city_norm.isin(target_cities.keys())
is_coop = df['brand'].fillna('').str.contains(
    r'\bcoop\b', case=False, na=False)

df_coop_cities = df.loc[is_target_city & is_coop].copy()
df_coop_cities['City'] = city_norm.map(target_cities)

coop_counts = df_coop_cities.groupby(
    'City', dropna=False).size().rename('Count').reset_index()
coop_counts
```

```python
# h
col_city = 'city' if 'city' in df.columns else 'addr:city'
col_housenumber = 'housenumber' if 'housenumber' in df.columns else 'addr:housenumber'
col_postcode = 'postcode' if 'postcode' in df.columns else 'addr:postcode'


def not_empty(series):
    return series.notna() & series.astype(str).str.strip().ne('')


mask_complete = (
    not_empty(df['brand']) &
    not_empty(df[col_city]) &
    not_empty(df[col_housenumber]) &
    not_empty(df[col_postcode])
)

df_complete_addr = df.loc[mask_complete].copy()
df_complete_addr.head()
```

```python
# i
if 'opening_hours' not in df.columns:
    if 'tags' in df.columns:
        df['opening_hours'] = df['tags'].apply(
            lambda t: t.get('opening_hours') if isinstance(t, dict) else None
        )
    else:
        # Falls dein DF schon flach ist, aber der Spaltenname anders lautet:
        for cand in ['opening_hours', 'addr:opening_hours']:
            if cand in df.columns:
                df['opening_hours'] = df[cand]
                break
```

```python
has_opening = df['opening_hours'].notna() & df['opening_hours'].astype(str).str.strip().ne('')
df_with_opening_hours = df.loc[has_opening].copy()
df_with_opening_hours.head()
```
